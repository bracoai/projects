```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick  # For setting exact numbers on the y-axis
from IPython.display import display, Markdown
import numpy as np
from statsmodels.stats.outliers_influence import variance_inflation_factor
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import LabelEncoder, StandardScaler

# Load the compressed CSV file
file_path = 'accepted_2007_to_2018q4.csv.gz'
data = pd.read_csv(file_path, low_memory=False)

# Limit the dataset to the first 1000 rows and save it to an excel file
#data_limited = data.head(1000)
#data_limited.to_excel('limited_dataset.xlsx', index=False)

# Check the shape of the dataset
rows, columns = data.shape
print(f"The initial dataset has {columns} columns (variables) and {rows} rows (observations per variable + 1).\n")

# Insight after analyzing the dataset size and structure
dataset_insight = """
<div style='color: blue;'>
    <b>Dataset Insight:</b>
    <ul>
        <li>After analyzing the size of the dataset, I found that it was too large to export directly to an Excel file due to the number of rows.</li>
        <li>I created a smaller (1000 rows) Excel file for a faster overview of the data structure.</li>
        <li>I identified the variable <b>loan_status</b>, which has strong potential for a default prediction model.</li>
        <li>Focusing on this variable will allow me to choose explanatory variables with purpose.</li>
    </ul>
</div>
"""
display(Markdown(dataset_insight))

# Ensure unique categories in loan_status
print("Loan Status Value Counts:")
data['loan_status'] = data['loan_status'].astype(str).str.strip()  # Remove extra spaces
loan_status_counts = data['loan_status'].value_counts()

# Print each value and its count
for value, count in loan_status_counts.items():
    print(f" {value}: {count:,}")  # Add comma formatting to the counts

# Function to plot bar chart with exact numbers on the y-axis
def plot_bar_chart(data, column, title, xlabel, xtick_labels=None):
    plt.figure(figsize=(8, 6))
    counts = data[column].value_counts().sort_index()

    ax = counts.plot(kind='bar', color='skyblue', edgecolor='black')
    plt.title(title, fontsize=16)
    plt.xlabel(xlabel, fontsize=12)
    plt.ylabel('Count', fontsize=12)

    # If xtick_labels is provided, use it for x-axis ticks
    if xtick_labels:
        plt.xticks(ticks=range(len(xtick_labels)), labels=xtick_labels, rotation=0, fontsize=10)
    else:
        plt.xticks(rotation=45, ha='right', fontsize=10, wrap=True)

    # Set exact numbers on the y-axis
    ax.yaxis.set_major_formatter(mtick.StrMethodFormatter('{x:,.0f}'))  # Format with commas

    plt.tight_layout()
    plt.show()

# Plot distribution of loan_status
plot_bar_chart(data, 'loan_status', 'Distribution of Loan Status', 'Loan Status')

# Insight after checking a potential dependent variable for a model
loan_status_insight = """
<div style='color: blue;'>
    <b>loan status insight:</b>
    <ul>
        <li>The observations are enough in quantity to discard data that will make some noise and still provide good insights.</li>
        <li>For example, there are more than 800,000 active loans, which can vary on age, some could be about to end and some just started.</li>
        <li>To avoid complexity on this assignment, I opted for simply divide the Loan Status variable into Default (Charged off) and Fully Paid.</li>
    </ul>
</div>
"""
display(Markdown(loan_status_insight))

# Create the second binary variable: "loan_status_deflt_only"
data['loan_status_deflt_only'] = data['loan_status'].apply(
    lambda x: 1 if x == "Charged Off" else (0 if x == "Fully Paid" else None)
)

# Drop rows with missing values for the new variable
data_cleaned_deflt_only = data.dropna(subset=['loan_status_deflt_only'])

# Get the number of observations after filtering for "loan_status_deflt_only"
rows_deflt_only = len(data_cleaned_deflt_only)

# Print the number of rows that will remain in "loan_status_deflt_only"
print(f"Number of observations in 'loan_status_deflt_only' after reshaping for important categories and dropping missing values: {rows_deflt_only}")

# Print the shape of the cleaned datasets (optional)
print(f"Shape of dataset with 'loan_status_deflt_only': {data_cleaned_deflt_only.shape}")

# Plot distribution of loan_status_deflt_only
plot_bar_chart(
    data_cleaned_deflt_only, 
    'loan_status_deflt_only', 
    'Distribution of loan_status_deflt_only', 
    'Default-Only Loan Status (0 = Fully Paid, 1 = Charged Off)', 
    xtick_labels=['0 (Fully Paid)', '1 (Charged Off)']
)

# Insight before proceeding to check missing values
check_missing_values_insight = """
<div style='color: blue;'>
    <b>Next Step (Checking for missing values) Insight:</b>
    <ul>
        <li>Before selecting variables for the model, I will analyze the dataset for missing values.</li>
        <li>Identifying the extent of missing data in each variable is crucial for making informed preprocessing decisions.</li>
        <li>By addressing missing values appropriately, I can ensure the integrity of the model training process.</li>
    </ul>
</div>
"""
display(Markdown(check_missing_values_insight))

# Proceed with missing value thresholds analysis for the cleaned data
missing_values = data_cleaned_deflt_only.isnull().sum()

# Variables with 0 missing values
zero_missing = missing_values[missing_values == 0]
print(f"\nVariables with 0 missing values ({len(zero_missing)}):")
if len(zero_missing) > 0:
    print("\n".join([f"{col}: {count}" for col, count in zero_missing.items()]))

# Define thresholds for missing value ranges
ranges = {
    "less than 5%": (0, int(0.05 * rows_deflt_only)),
    "5% to 10%": (int(0.05 * rows_deflt_only), int(0.10 * rows_deflt_only)),
    "10% to 20%": (int(0.10 * rows_deflt_only), int(0.20 * rows_deflt_only)),
    "20% to 30%": (int(0.20 * rows_deflt_only), int(0.30 * rows_deflt_only)),
    "30% to 50%": (int(0.30 * rows_deflt_only), int(0.50 * rows_deflt_only)),
    "50% to 100%": (int(0.50 * rows_deflt_only), rows_deflt_only)
}

# Iterate through each range, filter, and count variables
for label, (low, high) in ranges.items():
    filtered_missing_values = missing_values[(missing_values > low) & (missing_values <= high)].sort_values()
    total_variables = len(filtered_missing_values)
    
    # Print the summary for this range
    print(f"\nMissing values ({label}):")
    if total_variables > 0:
        print("\n".join([f"{col}: {count}" for col, count in filtered_missing_values.items()]))
    else:
        print("No variables in this range.")
    print(f"\nTotal number of variables with {label} missing: {total_variables}")

# Insight about variables without missing values
zero_missing_insight = """
<div style='color: blue;'>
    <b>Zero Missing Values Insight:</b>
    <ul>
        <li>We identified 47 variables without missing values, which creates a strong starting point for building the model.</li>
        <li>To demonstrate my ability to handle missing data, I plan to include at least one variable with less than 5% missing values and apply imputation techniques.</li>
        <li>It's important to note that while some variables might seem interesting, the lack of sufficient observations can lead to inaccuracies, even with synthetic data.</li>
        <li>Before training a potential the model, I must run a correlation matrix and Variance Inflation Factor (VIF) analysis to identify and remove multicollinear variables.</li>
        <li>These steps ensure the variables included in the model are both relevant and free of redundancy, improving its performance and interpretability.</li>
        <li>For simplicity, I'll pick 3 numerical and 3 categorical variables to address what is asked in this assignment</li>
    </ul>
</div>
"""
display(Markdown(zero_missing_insight))

#Correlation Analysis:
# Filter only numerical variables with 0 missing values
numerical_vars = data_cleaned_deflt_only[zero_missing.index].select_dtypes(include=['float64', 'int64'])

# Add the dependent variable back to the filtered dataset
numerical_vars['loan_status_deflt_only'] = data_cleaned_deflt_only['loan_status_deflt_only']

# Display numerical variables with 0 missing values
print(f"\nNumerical Variables with 0 Missing Values ({len(numerical_vars.columns)}):")
print(list(numerical_vars.columns))

# Insight about the selection
numerical_insight = """
<div style='color: blue;'>
    <b>Numerical Variables Insight:</b>
    <ul>
        <li>Only numerical variables with 0 missing values are considered for further analysis.</li>
        <li>Including missing values could distort the correlation matrix and VIF analysis.</li>
        <li>The selected variables provide a cleaner dataset for identifying potential explanatory variables.</li>
    </ul>
</div>
"""
display(Markdown(numerical_insight))

# Step 1: Correlation Matrix
correlation_matrix = numerical_vars.corr()
correlations_with_target = correlation_matrix['loan_status_deflt_only'].drop(labels='loan_status_deflt_only').sort_values(ascending=False)

# Display the correlations sorted by the highest correlation with the target variable
print("\nVariables Correlated with 'loan_status_deflt_only' (sorted by correlation):")
for var, corr in correlations_with_target.items():
    print(f"{var}: {corr:.2f}")

# Insight for correlation matrix
correlation_insight = """
<div style='color: blue;'>
    <b>Correlation Matrix Insight:</b>
    <ul>
        <li>The correlation matrix highlights variables most associated with <b>loan_status_deflt_only</b>, aiding in feature selection.</li>
        <li><b>recoveries</b> and <b>collection_recovery_fee</b> likely correlate highly as they result from defaults, not predict them.</li>
        <li>For further exploration, I selected <b>int_rate</b>, <b>last_fico_range_high</b>, and <b>last_pymnt_amnt</b> as numerical variables.</li>
        <li>Additionally, I will choose categorical variables for handling missing values, encoding, and scaling as part of the assignment.</li>
    </ul>
</div>
"""
display(Markdown(correlation_insight))

# Narrow down to selected columns
selected_columns = ["loan_status_deflt_only", "int_rate", "last_fico_range_high", 
                    "last_pymnt_amnt", "grade", "home_ownership", "application_type"]
data_narrowed = data_cleaned_deflt_only[selected_columns].copy()  # Ensure a copy is made

# Display summary of narrowed dataset
summary = data_narrowed.describe(include='all').transpose()
display(summary)

# Insight for narrowed dataset
narrowed_insight = """
<div style='color: blue;'>
    <b>Narrowed Dataset Insight:</b>
    <ul>
        <li>The dataset has been reduced to include only the target variable and selected predictors.</li>
        <li>3 numerical variables (<b>int_rate</b>, <b>last_fico_range_high</b>, <b>last_pymnt_amnt</b>) and 3 categorical variables (<b>grade</b>, <b>home_ownership</b>, <b>application_type</b>) are retained for further analysis.</li>
        <li>This ensures a manageable dataset for focused preprocessing and modeling.</li>
    </ul>
</div>
"""
display(Markdown(narrowed_insight))

# Handling missing values with mean imputation for numerical variables
imputer = SimpleImputer(strategy="mean")
data_narrowed.loc[:, ["int_rate", "last_fico_range_high", "last_pymnt_amnt"]] = imputer.fit_transform(
    data_narrowed[["int_rate", "last_fico_range_high", "last_pymnt_amnt"]]
)

# Calculate missing values before and after imputation
missing_before = data_narrowed[["int_rate", "last_fico_range_high", "last_pymnt_amnt"]].isnull().sum()
missing_after = data_narrowed[["int_rate", "last_fico_range_high", "last_pymnt_amnt"]].isnull().sum()

missing_table = pd.DataFrame({
    "Missing Before Imputation": missing_before,
    "Missing After Imputation": missing_after
})

display(missing_table)

# Insight for missing value handling
missing_insight = """
<div style='color: blue;'>
    <b>Missing Value Handling Insight:</b>
    <ul>
        <li>Missing values for numerical features were replaced with the mean using imputation.</li>
        <li>This ensures the dataset is complete and ready for further preprocessing.</li>
    </ul>
</div>
"""
display(Markdown(missing_insight))

# Encoding categorical variables
# Grade: Assign higher numbers to higher grades (A=7, ..., G=1)
grade_mapping = {"A": 7, "B": 6, "C": 5, "D": 4, "E": 3, "F": 2, "G": 1}
data_narrowed.loc[:, "grade"] = data_narrowed["grade"].map(grade_mapping)

# Home Ownership: Assign 3=OWN, 2=MORTGAGE, 1=RENT
home_ownership_mapping = {"OWN": 3, "MORTGAGE": 2, "RENT": 1}
data_narrowed.loc[:, "home_ownership"] = data_narrowed["home_ownership"].map(home_ownership_mapping)

# Application Type: Assign 1 for "Individual", 0 for "Joint App"
application_type_mapping = {"Individual": 1, "Joint App": 0}
data_narrowed.loc[:, "application_type"] = data_narrowed["application_type"].map(application_type_mapping)

# Frequency distribution of categorical variables before encoding
print("\nFrequency Distribution Before Encoding:")
print(data_cleaned_deflt_only[["grade", "home_ownership", "application_type"]].value_counts())

# Frequency distribution after encoding
print("\nFrequency Distribution After Encoding:")
print(data_narrowed[["grade", "home_ownership", "application_type"]].value_counts())

# Insight for categorical encoding
categorical_insight = """
<div style='color: blue;'>
    <b>Categorical Encoding Insight:</b>
    <ul>
        <li>Categorical variables have been successfully transformed into numerical values for analysis:</li>
        <li><b>grade:</b> Encoded from A=7 (highest) to G=1 (lowest).</li>
        <li><b>home_ownership:</b> Encoded from OWN=3 to RENT=1.</li>
        <li><b>application_type:</b> Encoded as Individual=1 and Joint App=0.</li>
    </ul>
</div>
"""
display(Markdown(categorical_insight))


# Scaling numerical features using StandardScaler
scaler = StandardScaler()
data_narrowed.loc[:, ["int_rate", "last_fico_range_high", "last_pymnt_amnt"]] = scaler.fit_transform(
    data_narrowed[["int_rate", "last_fico_range_high", "last_pymnt_amnt"]]
)

# Plot histograms before scaling
fig, axs = plt.subplots(1, 3, figsize=(18, 5))
data_cleaned_deflt_only[["int_rate", "last_fico_range_high", "last_pymnt_amnt"]].hist(ax=axs, bins=20, color='skyblue', edgecolor='black')
plt.suptitle("Numerical Features Before Scaling")
plt.tight_layout()
plt.show()

# Plot histograms after scaling
fig, axs = plt.subplots(1, 3, figsize=(18, 5))
data_narrowed[["int_rate", "last_fico_range_high", "last_pymnt_amnt"]].hist(ax=axs, bins=20, color='green', edgecolor='black')
plt.suptitle("Numerical Features After Scaling")
plt.tight_layout()
plt.show()

# Insight for scaling
scaling_insight = """
<div style='color: blue;'>
    <b>Scaling Insight:</b>
    <ul>
        <li>Numerical features have been standardized using <b>StandardScaler</b>.</li>
        <li>This ensures all features have a mean of 0 and standard deviation of 1, making them comparable for analysis.</li>
    </ul>
</div>
"""
display(Markdown(scaling_insight))

final_dataset_insight = """
<div style='color: blue;'>
    <b>Final Processed Dataset Insight:</b>
    <ul>
        <li>The final dataset is complete, with missing values handled, categorical variables encoded, and numerical features scaled.</li>
        <li>It includes 6 predictors and 1 target variable, ensuring a clean dataset for future modeling.</li>
        <li>Visualization and exploration confirm the dataset's readiness for predictive analysis.</li>
    </ul>
</div>
"""
display(Markdown(final_dataset_insight))
```
